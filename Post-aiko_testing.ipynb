```python
def norm_select(particles, groups, radius=0.01, samples=10, location=22669):
    galSelect = groupData.loc[location]  #location
    galPosX = galSelect["posX"]
    galPosY = galSelect["posY"]
    galPosZ = galSelect["posZ"]
    galRad = radius             #radius
    galSnap = galSelect["Snap"]
    galFiltered = (data.Snap == galSnap)&\
                  (data.posX<(galPosX+galRad))&(data.posX>(galPosX-galRad))&\
                  (data.posY<(galPosY+galRad))&(data.posY>(galPosY-galRad))&\
                  (data.posZ<(galPosZ+galRad))&(data.posZ>(galPosZ-galRad))

    ind = data[galFiltered].sample(samples).copy(deep=True)             #samples
    ind.posX = (ind.posX - galPosX)*1000.0
    ind.posY = (ind.posY - galPosY)*1000.0
    return ind
```

```python
def orbit(particles, groups, samples=10, interpol="cubic", steps=10, location=22669, radius=0.01, normal=True):
    ind = norm_select(particles, groups, radius, samples, location)
    #pd.apply research
    #ind.plot(x="posX",y="posY", kind="scatter")

    ivec=ind.index.drop_duplicates()
    normList = []
    for i in range(len(ivec)):
        tCords = data.loc[ivec[i]]
        tX = np.array(tCords["posX"])
        tY = np.array(tCords["posY"])
        tList = trrot(tX, tY)
        tdf = pd.DataFrame({"posX":tList[0], "posY":tList[1], "Ra":tList[2]})
        normList.append(tdf)

    if interpol == False:
        line = normList[0]*0
        ax =line.plot(x="posX",y="posY", legend=False)

        for i in normList:
            i.plot(x="posX",y="posY",ax=ax, legend=False, grid=True)

    if interpol == "cubic" or interpol == "quadratic":
        line = normList[0]*0                                #       if interpol=True
        ax =line.plot(x="posX",y="posY", legend=False)
        for i in normList:
            x = i["posX"]
            y = i["posY"]
            xSmooth = np.linspace(np.min(x), np.max(x), len(x)*steps+1)           #steps
            fspline = scipy.interpolate.interp1d(x, y, interpol)            #interpol
            ySmooth = fspline(xSmooth)
            ism = pd.DataFrame({"Xsmooth":xSmooth, "Ysmooth":ySmooth})
            ism.plot(x="Xsmooth", y="Ysmooth", ax=ax, legend=False, grid=True)
```
